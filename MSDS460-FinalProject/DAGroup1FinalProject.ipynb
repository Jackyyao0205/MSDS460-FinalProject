```python
import pulp
```

```python
# Define the MLP problem
mlp_programming = pulp.LpProblem("NU_Industries_Multi_Period_Production_Distribution", pulp.LpMaximize)
```

```python
# Decision Variable Declaration

# Production of Widget in factory A
Xwa1 = pulp.LpVariable("Xwa1", lowBound=0, cat='Continuous')
Xwa2 = pulp.LpVariable("Xwa2", lowBound=0, cat='Continuous')
Xwa3 = pulp.LpVariable("Xwa3", lowBound=0, cat='Continuous')
Xwa4 = pulp.LpVariable("Xwa4", lowBound=0, cat='Continuous')
Xwa5 = pulp.LpVariable("Xwa5", lowBound=0, cat='Continuous')

# Production of Gadgets in factory A
Xga1 = pulp.LpVariable("Xga1", lowBound=0, cat='Continuous')
Xga2 = pulp.LpVariable("Xga2", lowBound=0, cat='Continuous')
Xga3 = pulp.LpVariable("Xga3", lowBound=0, cat='Continuous')
Xga4 = pulp.LpVariable("Xga4", lowBound=0, cat='Continuous')
Xga5 = pulp.LpVariable("Xga5", lowBound=0, cat='Continuous')

# Production of Flugels in factory A
Xfa1 = pulp.LpVariable("Xfa1", lowBound=0, cat='Continuous')
Xfa2 = pulp.LpVariable("Xfa2", lowBound=0, cat='Continuous')
Xfa3 = pulp.LpVariable("Xfa3", lowBound=0, cat='Continuous')
Xfa4 = pulp.LpVariable("Xfa4", lowBound=0, cat='Continuous')
Xfa5 = pulp.LpVariable("Xfa5", lowBound=0, cat='Continuous')

# Production of Widget in factory B
Xwb1 = pulp.LpVariable("Xwb1", lowBound=0, cat='Continuous')
Xwb2 = pulp.LpVariable("Xwb2", lowBound=0, cat='Continuous')
Xwb3 = pulp.LpVariable("Xwb3", lowBound=0, cat='Continuous')
Xwb4 = pulp.LpVariable("Xwb4", lowBound=0, cat='Continuous')
Xwb5 = pulp.LpVariable("Xwb5", lowBound=0, cat='Continuous')

# Production of Gadgets in factory B
Xgb1 = pulp.LpVariable("Xgb1", lowBound=0, cat='Continuous')
Xgb2 = pulp.LpVariable("Xgb2", lowBound=0, cat='Continuous')
Xgb3 = pulp.LpVariable("Xgb3", lowBound=0, cat='Continuous')
Xgb4 = pulp.LpVariable("Xgb4", lowBound=0, cat='Continuous')
Xgb5 = pulp.LpVariable("Xgb5", lowBound=0, cat='Continuous')

# Production of Flugels in factory B
Xfb1 = pulp.LpVariable("Xfb1", lowBound=0, cat='Continuous')
Xfb2 = pulp.LpVariable("Xfb2", lowBound=0, cat='Continuous')
Xfb3 = pulp.LpVariable("Xfb3", lowBound=0, cat='Continuous')
Xfb4 = pulp.LpVariable("Xfb4", lowBound=0, cat='Continuous')
Xfb5 = pulp.LpVariable("Xfb5", lowBound=0, cat='Continuous')

# Inventory Storage of Widget in factory A
Iwa1 = pulp.LpVariable("Iwa1", lowBound=0, cat='Continuous')
Iwa2 = pulp.LpVariable("Iwa2", lowBound=0, cat='Continuous')
Iwa3 = pulp.LpVariable("Iwa3", lowBound=0, cat='Continuous')
Iwa4 = pulp.LpVariable("Iwa4", lowBound=0, cat='Continuous')

# Inventory Storage of Gadget in factory A
Iga1 = pulp.LpVariable("Iga1", lowBound=0, cat='Continuous')
Iga2 = pulp.LpVariable("Iga2", lowBound=0, cat='Continuous')
Iga3 = pulp.LpVariable("Iga3", lowBound=0, cat='Continuous')
Iga4 = pulp.LpVariable("Iga4", lowBound=0, cat='Continuous')

# Invetory Storage of Flugels in factory A
Ifa1 = pulp.LpVariable("Ifa1", lowBound=0, cat='Continuous')
Ifa2 = pulp.LpVariable("Ifa2", lowBound=0, cat='Continuous')
Ifa3 = pulp.LpVariable("Ifa3", lowBound=0, cat='Continuous')
Ifa4 = pulp.LpVariable("Ifa4", lowBound=0, cat='Continuous')

# Inventory Storage of Widget in factory B
Iwb1 = pulp.LpVariable("Iwb1", lowBound=0, cat='Continuous')
Iwb2 = pulp.LpVariable("Iwb2", lowBound=0, cat='Continuous')
Iwb3 = pulp.LpVariable("Iwb3", lowBound=0, cat='Continuous')
Iwb4 = pulp.LpVariable("Iwb4", lowBound=0, cat='Continuous')

# Inventory Storage of Gadget in factory B
Igb1 = pulp.LpVariable("Igb1", lowBound=0, cat='Continuous')
Igb2 = pulp.LpVariable("Igb2", lowBound=0, cat='Continuous')
Igb3 = pulp.LpVariable("Igb3", lowBound=0, cat='Continuous')
Igb4 = pulp.LpVariable("Igb4", lowBound=0, cat='Continuous')

# Invetory Storage of Flugels in factory B
Ifb1 = pulp.LpVariable("Ifb1", lowBound=0, cat='Continuous')
Ifb2 = pulp.LpVariable("Ifb2", lowBound=0, cat='Continuous')
Ifb3 = pulp.LpVariable("Ifb3", lowBound=0, cat='Continuous')
Ifb4 = pulp.LpVariable("Ifb4", lowBound=0, cat='Continuous')

# Cost of advertising of Widget in each time period
Aw1 = pulp.LpVariable("Aw1", lowBound=0, cat='Continuous')
Aw2 = pulp.LpVariable("Aw2", lowBound=0, cat='Continuous')
Aw3 = pulp.LpVariable("Aw3", lowBound=0, cat='Continuous')
Aw4 = pulp.LpVariable("Aw4", lowBound=0, cat='Continuous')

# Cost of advertising for Gadget in each time period
Ag1 = pulp.LpVariable("Ag1", lowBound=0, cat='Continuous')
Ag2 = pulp.LpVariable("Ag2", lowBound=0, cat='Continuous')
Ag3 = pulp.LpVariable("Ag3", lowBound=0, cat='Continuous')
Ag4 = pulp.LpVariable("Ag4", lowBound=0, cat='Continuous')

# Cost of advertising of Flugel in each time period
Af1 = pulp.LpVariable("Af1", lowBound=0, cat='Continuous')
Af2 = pulp.LpVariable("Af2", lowBound=0, cat='Continuous')
Af3 = pulp.LpVariable("Af3", lowBound=0, cat='Continuous')
Af4 = pulp.LpVariable("Af4", lowBound=0, cat='Continuous')

# Overtime labour used in Plant A in each time period
Oa1 = pulp.LpVariable("Oa1", lowBound=0, cat='Continuous')
Oa2 = pulp.LpVariable("Oa2", lowBound=0, cat='Continuous')
Oa3 = pulp.LpVariable("Oa3", lowBound=0, cat='Continuous')
Oa4 = pulp.LpVariable("Oa4", lowBound=0, cat='Continuous')
Oa5 = pulp.LpVariable("Oa5", lowBound=0, cat='Continuous')

# Overtime labour used in Plant B in each time period
Ob1 = pulp.LpVariable("Ob1", lowBound=0, cat='Continuous')
Ob2 = pulp.LpVariable("Ob2", lowBound=0, cat='Continuous')
Ob3 = pulp.LpVariable("Ob3", lowBound=0, cat='Continuous')
Ob4 = pulp.LpVariable("Ob4", lowBound=0, cat='Continuous')
Ob5 = pulp.LpVariable("Ob5", lowBound=0, cat='Continuous')

# Raw material used in pounds
RMa11 = pulp.LpVariable("RMa11", lowBound=0, cat='Continuous')
RMa12 = pulp.LpVariable("RMa12", lowBound=0, cat='Continuous')
RMa13 = pulp.LpVariable("RMa13", lowBound=0, cat='Continuous')
RMa14 = pulp.LpVariable("RMa14", lowBound=0, cat='Continuous')
RMa15 = pulp.LpVariable("RMa15", lowBound=0, cat='Continuous')

RMa21 = pulp.LpVariable("RMa21", lowBound=0, cat='Continuous')
RMa22 = pulp.LpVariable("RMa22", lowBound=0, cat='Continuous')
RMa23 = pulp.LpVariable("RMa23", lowBound=0, cat='Continuous')
RMa24 = pulp.LpVariable("RMa24", lowBound=0, cat='Continuous')
RMa25 = pulp.LpVariable("RMa25", lowBound=0, cat='Continuous')

RMb11 = pulp.LpVariable("RMb11", lowBound=0, cat='Continuous')
RMb12 = pulp.LpVariable("RMb12", lowBound=0, cat='Continuous')
RMb13 = pulp.LpVariable("RMb13", lowBound=0, cat='Continuous')
RMb14 = pulp.LpVariable("RMb14", lowBound=0, cat='Continuous')
RMb15 = pulp.LpVariable("RMb15", lowBound=0, cat='Continuous')

RMb21 = pulp.LpVariable("RMb21", lowBound=0, cat='Continuous')
RMb22 = pulp.LpVariable("RMb22", lowBound=0, cat='Continuous')
RMb23 = pulp.LpVariable("RMb23", lowBound=0, cat='Continuous')
RMb24 = pulp.LpVariable("RMb24", lowBound=0, cat='Continuous')
RMb25 = pulp.LpVariable("RMb25", lowBound=0, cat='Continuous')


Xwa = Xwa1 + Xwa2 + Xwa3 + Xwa4 + Xwa5
Xga = Xga1 + Xga2 + Xga3 + Xga4 + Xga5
Xfa = Xfa1 + Xfa2 + Xfa3 + Xfa4 + Xfa5

Xwb = Xwb1 + Xwb2 + Xwb3 + Xwb4 + Xwb5
Xgb = Xgb1 + Xgb2 + Xgb3 + Xgb4 + Xgb5
Xfb = Xfb1 + Xfb2 + Xfb3 + Xfb4 + Xfb5

Aw = Aw1 + Aw2 + Aw3 + Aw4
Ag = Ag1 + Ag2 + Ag3 + Ag4
Af = Af1 + Af2 + Af3 + Af4

all_decision_vars = [
    Xwa1, Xwa2, Xwa3, Xwa4, Xwa5,
    Xga1, Xga2, Xga3, Xga4, Xga5,
    Xfa1, Xfa2, Xfa3, Xfa4, Xfa5,
    Xwb1, Xwb2, Xwb3, Xwb4, Xwb5,
    Xgb1, Xgb2, Xgb3, Xgb4, Xgb5,
    Xfb1, Xfb2, Xfb3, Xfb4, Xfb5,
    Iwa1, Iwa2, Iwa3, Iwa4,
    Iga1, Iga2, Iga3, Iga4,
    Ifa1, Ifa2, Ifa3, Ifa4,
    Iwb1, Iwb2, Iwb3, Iwb4,
    Igb1, Igb2, Igb3, Igb4,
    Ifb1, Ifb2, Ifb3, Ifb4,
    Aw1, Aw2, Aw3, Aw4,
    Ag1, Ag2, Ag3, Ag4,
    Af1, Af2, Af3, Af4,
    Oa1, Oa2, Oa3, Oa4, Oa5,
    Ob1, Ob2, Ob3, Ob4, Ob5,
    RMa11, RMa12, RMa13, RMa14, RMa15,
    RMa21, RMa22, RMa23, RMa24, RMa25,
    RMb11, RMb12, RMb13, RMb14, RMb15,
    RMb21, RMb22, RMb23, RMb24, RMb25
]
```

```python
# Objective

# Revenue
revenue = 2490 * (Xwa + Xwb) + 1990 * (Xga + Xgb) + 2970 * (Xfa + Xfb)

# Costs

# Regula Labour costs
regular_labour_costs_A = 11 * (Oa1 + Oa2) + (11 * 1.05) * (Oa3 + Oa4 + Oa5)
regular_labour_costs_B = 11 * (Ob1 + Ob2) + (11 * 1.1) * (Ob3 + Ob4 + Ob5)
# Overtime Costs
overtime_costs_A = 16.50 * (Oa1 + Oa2) + (16.50 * 1.05) * (Oa3 + Oa4 + Oa5)
overtime_costs_B = 16.50 * (Ob1 + Ob2) + (16.50 * 1.1) * (Ob3 + Ob4 + Ob5)
# Total Labour Costs
total_labour_costs = regular_labour_costs_A + regular_labour_costs_B + overtime_costs_A + overtime_costs_B

# Inventory Costs
# Plant A
plant_a_widget_inventory_costs = 7.5 * Iwa1 + 7.5 * Iwa2 + 7.5 * Iwa3 + 7.5 * Iwa4
plant_a_gadget_inventory_costs = 5.5 * Iga1 + 5.5 * Iga2 + 5.5 * Iga3 + 5.5 * Iga4
plant_a_flugel_inventory_costs = 6.5 * Ifa1 + 6.5 * Ifa2 + 6.5 * Ifa3 + 6.5 * Ifa4
# Plant B
plant_b_widget_inventory_costs = 7.8 * Iwb1 + 7.8 * Iwb2 + 7.8 * Iwb3 + 7.8 * Iwb4
plant_b_gadget_inventory_costs = 5.7 * Igb1 + 5.7 * Igb2 + 5.7 * Igb3 + 5.7 * Igb4
plant_b_flugel_inventory_costs = 7 * Ifb1 + 7 * Ifb2 + 7 * Ifb3 + 7 * Ifb4
# Total Inventory Costs
total_inventory_costs = plant_a_widget_inventory_costs + plant_a_gadget_inventory_costs + plant_a_flugel_inventory_costs + plant_b_widget_inventory_costs + plant_b_gadget_inventory_costs + plant_b_flugel_inventory_costs

# Raw Material Costs
# Plant A
plant_a_raw_material_1_costs = 1.25 * RMa11 + 1.25 * RMa12 + 1.25 * RMa13 + 1.25 * RMa14 + 1.25 * RMa15
plant_a_raw_material_2_costs = 2.65 * RMa21 + 2.65 * RMa22 + 2.65 * RMa23 + 2.65 * RMa24 + 2.65 * RMa25
# Plant B
plant_b_raw_material_1_costs = 1.45 * RMb11 + 1.45 * RMb12 + 1.45 * RMb13 + 1.45 * RMb14 + 1.45 * RMb15
plant_b_raw_material_2_costs = 2.90 * RMb21 + 2.90 * RMb22 + 2.90 * RMb23 + 2.90 * RMb24 + 2.90 * RMb25
# Total Raw Material Costs
total_raw_material_costs = plant_a_raw_material_1_costs + plant_a_raw_material_2_costs

# Transportation Costs
# Plant A
plant_a_transportation_costs = 6.30 * Xwa + 4.60 * Xga + 5.50 * Xfa
# Plant B
plant_b_transportation_costs = 6.50 * Xwb + 5.0 * Xgb + 5.7 * Xfb
# Total Transportation Costs
total_transportation_costs = plant_a_transportation_costs + plant_b_transportation_costs

mlp_programming += revenue - total_labour_costs - total_inventory_costs - total_raw_material_costs - total_transportation_costs, "Objective"
```

```python
# Constraints

# Raw Material Constraints
# Amount used in each is less than amount imported
mlp_programming += 194*Xwa1 + 230*Xga1 + 178*Xfa1 <= RMa11, "Constraint_1"
mlp_programming += 188*Xwb1 + 225*Xgb1 + 170*Xfb1 <= RMb11, "Constraint_2"
mlp_programming += 194*Xwa2 + 230*Xga2 + 178*Xfa2 <= RMa12, "Constraint_3"
mlp_programming += 188*Xwb2 + 225*Xgb2 + 170*Xfb2 <= RMb12, "Constraint_4"
mlp_programming += 194*Xwa3 + 230*Xga3 + 178*Xfa3 <= RMa13, "Constraint_5"
mlp_programming += 188*Xwb3 + 225*Xgb3 + 170*Xfb3 <= RMb13, "Constraint_6"
mlp_programming += 194*Xwa4 + 230*Xga4 + 178*Xfa4 <= RMa14, "Constraint_7"
mlp_programming += 188*Xwb4 + 225*Xgb4 + 170*Xfb4 <= RMb14, "Constraint_8"
mlp_programming += 194*Xwa5 + 230*Xga5 + 178*Xfa5 <= RMa15, "Constraint_9"
mlp_programming += 188*Xwb5 + 225*Xgb5 + 170*Xfb5 <= RMb15, "Constraint_10"

mlp_programming += 8.6*Xwa1 + 11.6*Xfa1 <= RMa21, "Constraint_65"
mlp_programming += 9.2*Xwb1 + 10.8*Xfb1 <= RMb21, "Constraint_66"
mlp_programming += 8.6*Xwa2 + 11.6*Xfa2 <= RMa22, "Constraint_67"
mlp_programming += 9.2*Xwb2 + 10.8*Xfb2 <= RMb22, "Constraint_68"
mlp_programming += 8.6*Xwa3 + 11.6*Xfa3 <= RMa23, "Constraint_69"
mlp_programming += 9.2*Xwb3 + 10.8*Xfb3 <= RMb23, "Constraint_70"
mlp_programming += 8.6*Xwa4 + 11.6*Xfa4 <= RMa24, "Constraint_71"
mlp_programming += 9.2*Xwb4 + 10.8*Xfb4 <= RMb24, "Constraint_72"
mlp_programming += 8.6*Xwa5 + 11.6*Xfa5 <= RMa25, "Constraint_73"
mlp_programming += 9.2*Xwb5 + 10.8*Xfb5 <= RMb25, "Constraint_74"


# Labour Constraints
mlp_programming += 9.5*Xwa1 + 7.1*Xga1 + 11.1*Xfa1 <= 2500 + Oa1, "Constraint_11"
mlp_programming += 9.5*Xwa2 + 7.1*Xga2 + 11.1*Xfa2 <= 2500 + Oa2, "Constraint_12"
mlp_programming += 9.5*Xwa3 + 7.1*Xga3 + 11.1*Xfa3 <= 2500 + Oa3, "Constraint_13"
mlp_programming += 9.5*Xwa4 + 7.1*Xga4 + 11.1*Xfa4 <= 2500 + Oa4, "Constraint_14"
mlp_programming += 9.5*Xwa5 + 7.1*Xga5 + 11.1*Xfa5 <= 2500 + Oa5, "Constraint_15"

mlp_programming += 9.1*Xwb1 + 7.8*Xgb1 + 10.6*Xfb1 <= 3800 + Ob1, "Constraint_16"
mlp_programming += 9.1*Xwb2 + 7.8*Xgb2 + 10.6*Xfb2 <= 3800 + Ob2, "Constraint_17"
mlp_programming += 9.1*Xwb3 + 7.8*Xgb3 + 10.6*Xfb3 <= 3800 + Ob3, "Constraint_18"
mlp_programming += 9.1*Xwb4 + 7.8*Xgb4 + 10.6*Xfb4 <= 3800 + Ob4, "Constraint_19"
mlp_programming += 9.1*Xwb5 + 7.8*Xgb5 + 10.6*Xfb5 <= 3800 + Ob5, "Constraint_20"

# Amount of Raw Materials imported less than 70 tons (RM1) and 2.5 tons (RM2)
mlp_programming += RMa11 + RMb11 <= 140000, "Constraint_55"
mlp_programming += RMa12 + RMb12 <= 140000, "Constraint_56"
mlp_programming += RMa13 + RMb13 <= 140000, "Constraint_57"
mlp_programming += RMa14 + RMb14 <= 140000, "Constraint_58"
mlp_programming += RMa15 + RMb15 <= 140000, "Constraint_59"

mlp_programming += RMa21 + RMb21 <= 5000, "Constraint_60"
mlp_programming += RMa22 + RMb22 <= 5000, "Constraint_61"
mlp_programming += RMa23 + RMb23 <= 5000, "Constraint_62"
mlp_programming += RMa24 + RMb24 <= 5000, "Constraint_63"
mlp_programming += RMa25 + RMb25 <= 5000, "Constraint_64"

# Demand Satisfaction Constraints
mlp_programming += Xwa1 + Xwb1 - Iwa1 - Iwb1 >= 70, "Constraint_21"
mlp_programming += Xga1 + Xgb1 - Iga1 - Igb1 >= 200, "Constraint_22"
mlp_programming += Xfa1 + Xfb1 - Ifa1 - Ifb1 >= 140, "Constraint_23"

# The reason I am changing the constraints is bc division is not possible in pulp
mlp_programming += 160*(Xwa2 + Xwb2 + Iwa1 + Iwb1 - Iwa2 - Iwb2) >= 160 * 125 + Aw1, "Constraint_24"
mlp_programming += 120*(Xga2 + Xgb2 + Iga1 + Igb1 - Iga2 - Igb2) >= 120 * 300 + Ag1, "Constraint_25"
mlp_programming += 180*(Xfa2 + Xfb2 + Ifa1 + Ifa1 - Ifa2 - Ifb2) >= 180 * 175 + Af1, "Constraint_26"
mlp_programming += 160*(Xwa3 + Xwb3 + Iwa2 + Iwb2 - Iwa3 - Iwb3) >= 160 * 185 + Aw2, "Constraint_27"
mlp_programming += 120*(Xga3 + Xgb3 + Iga2 + Igb2 - Iga3 - Igb3) >= 120 * 295 + Ag2, "Constraint_28"
mlp_programming += 180*(Xfa3 + Xfb3 + Ifa2 + Ifb2 - Ifa3 - Ifb3) >= 180 * 205 + Af2, "Constraint_29"
mlp_programming += 160*(Xwa4 + Xwb4 + Iwa3 + Iwb3 - Iwa4 - Iwb4) >= 160 * 190 + Aw3, "Constraint_30"
mlp_programming += 120*(Xga4 + Xgb4 + Iga3 + Igb3 - Iga4 - Igb4) >= 120 * 245 + Ag3, "Constraint_31"
mlp_programming += 180*(Xfa4 + Xfb4 + Ifa3 + Ifb3 - Ifa4 - Ifb4) >= 180 * 235 + Af3, "Constraint_32"
mlp_programming += 160*(Xwa5 + Xwb5 + Iwa4 + Iwb4) >= 160 * 200 + Aw4, "Constraint_33"
mlp_programming += 120*(Xga5 + Xgb5 + Iga4 + Igb4) >= 120 * 240 + Ag4, "Constraint_34"
mlp_programming += 180*(Xfa5 + Xfb5 + Ifa4 + Ifb4) >= 180 * 230 + Af4, "Constraint_35"

# Each inventory can only store what that factory produced
mlp_programming += Iwa1 <= Xwa1, "Constraint_37"
mlp_programming += Iga1 <= Xga1, "Constraint_38"
mlp_programming += Ifa1 <= Xfa1, "Constraint_40"
mlp_programming += Iwb1 <= Xwb1, "Constraint_41"
mlp_programming += Igb1 <= Xgb1, "Constraint_43"
mlp_programming += Ifb1 <= Xfb1, "Constraint_44"
mlp_programming += Iwa2 <= Xwa2 + Iwa1, "Constraint_47"
mlp_programming += Iga2 <= Xga2 + Iga1, "Constraint_49"
mlp_programming += Ifa2 <= Xfa2 + Ifa1, "Constraint_50"
mlp_programming += Iwb2 <= Xwb2 + Iwb1, "Constraint_53"
mlp_programming += Igb2 <= Xgb2 + Igb1, "Constraint_94"
mlp_programming += Ifb2 <= Xfb2 + Ifb1, "Constraint_82"
mlp_programming += Iwa3 <= Xwa3 + Iwa2, "Constraint_83"
mlp_programming += Iga3 <= Xga3 + Iga2, "Constraint_84"
mlp_programming += Ifa3 <= Xfa3 + Ifa2, "Constraint_85"
mlp_programming += Iwb3 <= Xwb3 + Iwb2, "Constraint_86"
mlp_programming += Igb3 <= Xgb3 + Igb2, "Constraint_87"
mlp_programming += Ifb3 <= Xfb3 + Ifb2, "Constraint_88"
mlp_programming += Iwa4 <= Xwa4 + Iwa3, "Constraint_89"
mlp_programming += Iga4 <= Xga4 + Iga3, "Constraint_90"
mlp_programming += Ifa4 <= Xfa4 + Ifa3, "Constraint_91"
mlp_programming += Iwb4 <= Xwb4 + Iwb3, "Constraint_92"
mlp_programming += Igb4 <= Xgb4 + Igb3, "Constraint_93"
mlp_programming += Ifb4 <= Xfb4 + Ifb3, "Constraint_81"

# Inventory Constraints
mlp_programming += Iwa1 + Iga1 + Ifa1 <= 70, "Constraint_36"
mlp_programming += Iwa2 + Iga2 + Ifa2 <= 70, "Constraint_39"
mlp_programming += Iwa3 + Iga3 + Ifa3 <= 70, "Constraint_42"
mlp_programming += Iwa4 + Iga4 + Ifa4 <= 70, "Constraint_45"

mlp_programming += Iwb1 + Igb1 + Ifb1 <= 50, "Constraint_46"
mlp_programming += Iwb2 + Igb2 + Ifb2 <= 50, "Constraint_48"
mlp_programming += Iwb3 + Igb3 + Ifb3 <= 50, "Constraint_51"
mlp_programming += Iwb4 + Igb4 + Ifb4 <= 50, "Constraint_52"


# Advertising Constraint
mlp_programming += 160*Aw + 120*Ag + 180*Af <= 70000, "Constraint_54"
```

```python
# Step 6: Print the solution status and variable values
pulp.LpSolverDefault.msg = 1  # Enable solver messages
status = mlp_programming.solve(pulp.PULP_CBC_CMD(msg=True))
print("Status:", pulp.LpStatus[mlp_programming.status])
for decision_var in all_decision_vars:
    print(decision_var.name, "=", decision_var.varValue)
print("Objective value =", pulp.value(mlp_programming.objective))

# try:
#     with open("sensitivity.txt", "r") as f:
#         print("\n--- Sensitivity Analysis ---\n")
#         for line in f.readlines():
#             print(line.strip())
# except FileNotFoundError:
#     print("Sensitivity analysis file not found.")
```
